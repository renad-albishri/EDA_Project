# DIGITAL MEDIA COMPANY(DMC) FOR ADVERTISING AND ANNOUNCING

# Goal
The goal of this project is preprocessing and analyzing MTA Turnstile dataset to be used in identifying the most crowded subway stations to be choosen by the DMC to increase the profit of it.

# Prepossessing  
Cleaned the data by renaming columns, adding calculated fields, removing erroneous data, and aggregating by station and date.
it's done by added the 'Week_days,' 'Month,' and Year columns to be used for finding the highest traffic per months or days. The analysis found that some station weekdays had the most increased traffic, but this was less the case for some stations; certain days also had anomalous traffic flows.

To start exploring the goal, I used in this graph a bar chart with two features (Stations, Dialy_Entries) to show the most crowded subway stations wich are used to give many choices to the company which benefit from this data.

in this graph , I used a line chart as a subplot for two plots , the first plot where in the (first row, first column) is describe the total traffic per months and we notice that in Dec and July the most traffic and the low traffic in May.
the second plot where in the (first row, second column) is describe the total traffic per days and we notice that in Thursday the most traffic and the low traffic in Saturday.
